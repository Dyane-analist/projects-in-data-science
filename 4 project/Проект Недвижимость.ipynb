### Проверка и заполнение пропущенных значений

Определим, в каких колонках есть пропущенных значения, и сколько их в % отношении.

```python
(df.isna().mean()*100).round(2) # проверка % пропущенных значений в колонках
```

Обработаем и заполним пропущенные значения по порядку.

Поскольку данные столбца **'ceiling_height'** понадобятся нам в дальнейшем исследовании, заполним пропущенные значения в колонке с высотой потолков медианным значением по данному столбцу; медианное значение используем, чтобы избежать влияния выбросов на результат. 

Отсутствие данных в этом столбце может быть связано с ошибками выгрузки или отсутствием этой информации в объявлении, например, по причине того, что дом относится к типовым с высотой потолков 2,5-2,7 м, и автор объявления не счел информацию о высоте потолков релевантной для продажи квартиры.

```python
# заполнение пропущенных значений в колонке с высотой потолков медианным значением по данному столбцу 
df['ceiling_height'] = df['ceiling_height'].fillna(df['ceiling_height'].median()) 
```

Поскольку данные столбца **'floors_total'** понадобятся нам в дальнейшем исследовании, заполним пропущенные значения в колонке с количеством этажей в здании тем этажом, который указан в колонке "floor" для квартиры из объявления, т.к. в колонке "floor" пропусков нет, и в здании в любом случае не может быть этажей меньше, чем этаж квартиры в этом же здании. 

Отсутствие данных в этом столбце может быть связано с ошибками выгрузки или отсутствием этой информации в объявлении, т.к. автор объявления не счел информацию о количестве этажей в доме релевантной для продажи квартиры. 

```python
# заполнение пропущенных значений в колонке с количеством этажей в здании этажом квартиры из объявления
df['floors_total'] = df['floors_total'].fillna(df['floor'])
```

Поскольку данные столбца **'living_area'** понадобятся нам в дальнейшем исследовании, заполним пропущенные значения в клонке с жилой площадью усредненным значением жилой площади. Для этого найдем отношение средней жилой площади к средней общей площади по всему датафрейму, а затем заполним пропуски произведением этого отношения на общую площадь (в колонке с общей площадью пропусков нет).

Отсутствие данных в этом столбце может быть связано с ошибками выгрузки или отсутствием этой информации в объявлении, т.к. автор объявления не счел информацию об жилой площади в доме релевантной для продажи квартиры.

```python
# расчет отношения средней жилой площади к средней общей площади по датафрейму
living_ratio = df['living_area'].mean()/df['total_area'].mean()
# заполнение пропусков произведением этого отношения и общей площади
df['living_area'].fillna(df['total_area'] * living_ratio, inplace=True)
```

Колонка **"is_apartment"** не используется в данном исследовании, поэтому ее включаем в исследование без обработки пропусков.

Поскольку данные столбца **'kitchen_area'** понадобятся нам в дальнейшем исследовании, аналогично жилой площади выше заполним пропущенные значения в клонке с площадью кухни усредненным значением площади кухни. Для этого найдем отношение средней площади кухни к средней общей площади по всему датафрейму, а затем заполним пропуски произведением этого отношения на общую площадь (в колонке с общей площадью пропусков нет).

Отсутствие данных в этом столбце может быть связано с ошибками выгрузки или отсутствием этой информации в объявлении, т.к. автор объявления не счел информацию о количестве этажей в доме релевантной для продажи квартиры.

```python
# расчет отношения площади кухни к средней общей площади по датафрейму
kitchen_ratio = df['kitchen_area'].mean()/df['living_area'].mean()
# заполнение пропусков произведением этого отношения и общей площади
df['kitchen_area'].fillna(df['living_area'] * kitchen_ratio, inplace=True)
```

```python
# проверка площадь кухни + жилая площадь < общая площадь
df.query('kitchen_area + living_area > total_area')
```

```python
# замена превышений на total_area - living_area
df.loc[df['kitchen_area'] + df['living_area'] > df['total_area'], 'kitchen_area'] = df['total_area'] - df['living_area']  
```

```python
# проверка результата
df.query('kitchen_area + living_area > total_area')
```

Поскольку, вероятно, отсутствие информации о количестве балконов в колонке **'balcony'** говорит об отсутствии балконов в принципе, заменим пропущенные значения в колонке 'balcony' на 0. 

Эти данные нам в исследовании не потребуются, однако замена значений на 0 была оговорена в условии задания на исследование, поэтому проведем данную замену.

```python
df['balcony'] = df['balcony'].fillna(0) # замена пропусков на 0
```

Названия населенных пунктов в колонке **'locality_name'** понадобятся нам в дальнейшем исследовании, однако для их получения требуется провести отдельное исследование, не входящее в рамки текущего. Поэтому пропущенные значения в названии населенного пункта заменим пометкой об отсутствии этой информации 'no info'. В реальном исследовании для решения этой задачи нужно было бы сделать срез по отсутствующим значениям и попробовать поискать населенные пункты по адресам объектов (улица, дом, квартира).

Причина отсутствия названий в 0,21% объявлений, скорее всего - ошибка выгрузки или ее автоматического обогащения (если название населенного пункта было добавлено в датасет автоматически).

```python
# замена пропусков в названии населенного пункта пометкой об отсутствии этой информации
df['locality_name'] = df['locality_name'].fillna('no info') 
```

Данные колонки **"days_exposition"** понадобятся нам в дальнейшем исследовании. Вероятная причина отсутствия данных - ошибка выгрузки или отсутствие информации о дате снятия объявления с публикации. Поскольку очевидного способа заполнить эти пропуски нет, оставим колонку без изменений.

Колонки **"airports_nearest", "parks_around3000", "parks_nearest", "ponds_around3000", "ponds_nearest"** не используются в данном исследовании, поэтому их включаем в исследование без обработки пропусков.
В случае, если бы они использовались в исследовании, потребовалось бы сгруппировать данные и заполнить пропуски средними или медианными значениями.

Причина отсутствия данных по расстояниям, скорее всего - ошибка автоматического обогащения датасета (эти данные были добавлены автоматически).

В колонке **"city_centers_nearest"** данных нет по четверти объявлений, и эти данные понадобятся нам в дальнейшем исследовании. 

Причина отсутствия данных, скорее всего - ошибка автоматического обогащения выгрузки картографическими данными.

Заполним пропуски медианным значением по соответствующему населенному пункту, чтобы избежать влияния выбросов.

Сформируем список населенных пунктов с пропусками в колонке 'city_centers_nearest'.

```python
# формирование таблицы со списком населенных пунктов с пропусками в колонке 'city_centers_nearest'
locality_no_city_centers = df[df['city_centers_nearest'].isna()].groupby('locality_name')['locality_name'].count()
locality_no_city_centers.head() # вывод на экран первых 5 строк новой таблицы
```

Для выбора способа замены пропущенных значений оценим размер получившейся выборки.

```python
locality_no_city_centers.shape # оценка размеров выборки по количеству строк
```

Используем метод transform для замены пропущенных значений: заменим их медианным значением 'city_centers_nearest' по каждому населенному пункту, где есть пропуски.

```python
# замена пропусков среднеарифметическим значением по конкретному населенному пункту
df['city_centers_nearest'] = df['city_centers_nearest'].fillna(df.groupby('locality_name')['city_centers_nearest'].transform('median'))
```

Проверим корректность заполнения пропусков.
